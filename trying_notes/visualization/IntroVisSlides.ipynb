# Example of Text to Vectors

```python
import PlaYnlp.tokenizer as tkr
import PlaYnlp.vectorizer as vcr
```

```python
print "~~~~~~~~~~~~~~~~~~~~~"
print "/ ".join(vcr.tokenize_gen(vcr.jieba.cut)(u"今天天氣很好"))
print "~~~~~~~~~~~~~~~~~~~~~"
print "/ ".join(vcr.tokenize_gen(vcr.jieba.cut)(u"<p>今天天氣很好</p>"))
print "~~~~~~~~~~~~~~~~~~~~~"
print "/ ".join(vcr.tokenize_gen(lambda text:vcr.jieba.cut(vcr.nltk.clean_html(text)))(u"<p>今天天氣很好</p>"))
print "~~~~~~~~~~~~~~~~~~~~~"

jieba_without_html_tokenizer = vcr.tokenize_gen(lambda text:vcr.jieba.cut(vcr.nltk.clean_html(text)))
unigram_without_html_tokenizer = vcr.tokenize_gen(lambda text:tkr.ngram(vcr.nltk.clean_html(text),n=1))
bigram_without_html_tokenizer = vcr.tokenize_gen(lambda text:tkr.ngram(vcr.nltk.clean_html(text),n=2))

print "/ ".join(jieba_without_html_tokenizer(u"<p>今天天氣很好</p>"))
print "/ ".join(unigram_without_html_tokenizer(u"<p>今天天氣很好</p>"))
print "/ ".join(bigram_without_html_tokenizer(u"<p>今天天氣很好</p>"))
```

# Text Vectorizer in Scikit Learn

- ### [CountVectorizer](http://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html)
- ### [TfidfVectorizer](http://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html)

```python
jieba_vec_count_kwargs = {"tokenizer":jieba_without_html_tokenizer,"lowercase":False}
unigram_vec_count_kwargs = {"tokenizer":unigram_without_html_tokenizer,"lowercase":False}
bigram_vec_count_kwargs = {"tokenizer":bigram_without_html_tokenizer,"lowercase":False}
```

# Demo Data:

```python
import pandas as pd
```

```python
test_text_df = pd.DataFrame([u"今天天氣很好",u"今天天氣很爛",u"我恨它",u"它恨我",u"我愛它",u"它愛我",u"今天很衰","日子一天一天過","天天刷牙洗臉"])
test_text_df.columns = ["text"]
test_text_df
```

# Text to Vector ... 

```python
jieba_text2vec = vcr.vectorize_text(df=test_text_df,
                                      colname="text",
                                      query={}, 
                                      vect_gen=vcr.CountVectorizer, 
                                      vect_gen_init_kwargs = jieba_vec_count_kwargs)


unigram_text2vec = vcr.vectorize_text(df=test_text_df,
                                      colname="text",
                                      query={}, 
                                      vect_gen=vcr.CountVectorizer, 
                                      vect_gen_init_kwargs = unigram_vec_count_kwargs)

bigram_text2vec = vcr.vectorize_text(df=test_text_df,
                                      colname="text",
                                      query={}, 
                                      vect_gen=vcr.CountVectorizer, 
                                      vect_gen_init_kwargs = bigram_vec_count_kwargs)
```

# Solving of Optimization Problems:

# Linear Programming:

## $$ \begin{eqnarray} \min_{x_1,x_2} x_1 & + & x_2 \\ 2x_1 & + & x_2 & \geq 1 \\ x_1 & + & 3x_2 & \geq 1 \\ x_1 &&& \geq 0 \\ x_2 &&& \geq 0 \\ \end{eqnarray}$$

![](LP_Demo.png)

# Min-Max Problem:

## $$ \begin{eqnarray} \min_{x_1,x_2} \max( x_1 & , & x_2 ) \\ 2x_1 & + & x_2 & \geq 1 \\ x_1 & + & 3x_2 & \geq 1 \\ x_1 &&& \geq 0 \\ x_2 &&& \geq 0 \\ \end{eqnarray}$$


![](MinMax_Demo.png)

# Quadratic Programming:

## $$ \begin{eqnarray} \min_{x_1,x_2}  x_1^2 & + & 9x_2^2  \\ 2x_1 & + & x_2 & \geq 1 \\ x_1 & + & 3x_2 & \geq 1 \\ x_1 &&& \geq 0 \\ x_2 &&& \geq 0 \\ \end{eqnarray}$$
