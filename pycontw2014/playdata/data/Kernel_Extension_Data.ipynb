```python
import os 
import Orange

def convert_dataframe_to_orange(df, with_index=False, index_label=None):
    if with_index:
        df.to_csv('_temp_.csv',index = with_index, index_label=index_label)
    else:
        df.to_csv('_temp_.csv',index = with_index)
        
    orange_table = Orange.data.Table('_temp_.csv')
    os.unlink('_temp_.csv')
    return orange_table
```

```python
import pandas as pd 
```

```python
kernel_example_df = pd.DataFrame([["A",1.0,0.0],
                                  ["A",0.0,1.0],
                                  ["A",0.0,-1.0],
                                  ["B",-1.0,0.0],
                                  ["B",0.0,2.0],
                                  ["B",0.0,-2.0],
                                  ["B",-2.0,0.0]], columns=["L","x1","x2"])

kernel_example_df
```

```python
orange_dt = convert_dataframe_to_orange(kernel_example_df)
orange_dt.save("kernel_example_origin.tab")
```

```python
kernel_example_df["x1_2"] = kernel_example_df["x1"]**2
kernel_example_df["x2_2"] = kernel_example_df["x2"]**2
kernel_example_df["x12"] = kernel_example_df["x2"]*kernel_example_df["x1"]
kernel_example_df
```

```python
orange_dt = convert_dataframe_to_orange(kernel_example_df)
orange_dt.save("kernel_example_extend.tab")
```
