```python
import pandas as pd
```

```python
corpus_df = pd.DataFrame([["Movie","a perfect world"],
                          ["Movie","my perfect woman"],
                          ["Movie","pretty woman"],
                          ["Song","a perfect day"],
                          ["Song","electric storm"],
                          ["Song","mother rainy day"]], columns=["class","text"])
corpus_df
```

```python
import PlaYnlp.tokenizer as tkr 
```

```python
ngram_data_list = []
for xx in corpus_df.values.tolist():
    for zz in map(lambda yy:[xx[0]]+yy,list(tkr.ngram(xx[1].split(),n=1))):
        ngram_data_list.append(zz)

unigram_df = pd.DataFrame(ngram_data_list,columns = ["class","w"])

ngram_data_list = []
for xx in corpus_df.values.tolist():
    for zz in map(lambda yy:[xx[0]]+yy,list(tkr.ngram(xx[1].split(),n=2))):
        ngram_data_list.append(zz)
bigram_df = pd.DataFrame(ngram_data_list,columns = ["class","w1","w2"])
```

```python
unigram_df
```

# Compute P(w) ... 

```python
unigram_df.groupby(["w"]).size().to_frame()
```

```python
unigram_df.groupby(["w"]).size() / unigram_df.groupby(["w"]).size().sum()
```

# Compute P(class,w) ... 

```python
(unigram_df.groupby(["class","w"]).size() / unigram_df.groupby(["class","w"]).size().sum()).unstack().fillna(0)
```

```python
unigram_count = unigram_df.groupby(["class","w"]).size().unstack().fillna(0)
unigram_count
```

```python
unigram_count.sum(1).to_frame()
```

```python
unigram_count.join(unigram_count.sum(1).to_frame(),how="inner")
```

# Laplace Smoothing ... 

```python
k = 1
s = unigram_df.groupby(["w"]).size().to_frame().shape[0]
print "s = ",s
(unigram_count + k).join(unigram_count.sum(1).to_frame() + s*k,how="inner")
```

# 從一個向量化的範例 ... 思考「距離」的本質？ ...

```python
import pandas as pd 
test_text_df = pd.DataFrame([u"今天天氣很好",u"今天天氣很爛",u"我恨它",u"它恨我",u"我愛它",u"它愛我",u"今天很衰",u"日子一天一天過",u"天天刷牙洗臉"])
test_text_df.columns = ["text"]
test_text_df
```

```python
import PlaYnlp.tokenizer as tkr 
import PlaYnlp.vectorizer as vcr 
import numpy as np
import scipy as sp
jieba_without_html_tokenizer = tkr.tokenize_gen(lambda text:tkr.jieba.cut(tkr.nltk.clean_html(text)))
unigram_without_html_tokenizer = tkr.tokenize_gen(lambda text:tkr.ngram(tkr.nltk.clean_html(text),n=1))
bigram_without_html_tokenizer = tkr.tokenize_gen(lambda text:tkr.ngram(tkr.nltk.clean_html(text),n=2))
jieba_vec_count_kwargs = {"tokenizer":jieba_without_html_tokenizer,"lowercase":False}
unigram_vec_count_kwargs = {"tokenizer":unigram_without_html_tokenizer,"lowercase":False}
bigram_vec_count_kwargs = {"tokenizer":bigram_without_html_tokenizer,"lowercase":False}
```

```python
test_jieba_sdtm = vcr.vectorize_text(df=test_text_df,
                                     text_col="text", 
                                     idx_col=None, 
                                     cond_query={},
                                     idx_query= [],
                                     vect_gen=vcr.CountVectorizer, 
                                     vect_gen_init_kwargs = jieba_vec_count_kwargs)

test_unigram_sdtm = vcr.vectorize_text(df=test_text_df,
                                     text_col="text", 
                                     idx_col=None, 
                                     cond_query={},
                                     idx_query= [],
                                     vect_gen=vcr.CountVectorizer, 
                                     vect_gen_init_kwargs = unigram_vec_count_kwargs)

test_bigram_sdtm = vcr.vectorize_text(df=test_text_df,
                                     text_col="text", 
                                     idx_col=None, 
                                     cond_query={},
                                     idx_query= [],
                                     vect_gen=vcr.CountVectorizer, 
                                     vect_gen_init_kwargs = bigram_vec_count_kwargs)
```

```python
test_unigram_sdtm.to_pandas_df
```

```python
test_bigram_sdtm.to_pandas_df
```

```python
test_jieba_sdtm.to_pandas_df
```
